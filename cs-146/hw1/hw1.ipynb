```python
X, y = np.array_split(df31.to_numpy(), (2,), axis=1)
model = Perceptron(max_iter = 1000)
model.fit(X, y.ravel())
```

```python
w = model.coef_[0]
b = model.intercept_
u = model.t_
print("Weights: ", w)
print("Bias: ", b)
print("Updates: ", u)
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(2, 5)
ax.scatter(df31[0], df31[1], c=df31[2])
ax.plot(x, - (b + w[0]*x) / w[1])
plt.title("2031 Admissions")
plt.savefig("5bii.png")
```

```python
def shortest_distance(x, y): 
    return abs((4.53609 * x + 2.42091 * y - 22)) / (math.sqrt(4.53609 * 4.53609 + 2.42091 * 2.42091))
```

```python
best = []
minDist = 100
for row in X:
    x = row[0]
    y = row[1]
    dist = shortest_distance(x, y)
    if (dist < minDist):
        best = row
        minDist = dist
        print(minDist)
```
